```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import os
import glob
import torch
```

```python
#####LSTM DTW PLOT

os.chdir("/Users/alainamahalanabis/Documents/cycleGAN/missing_V_0.1")
dtw_normal_files = glob.glob('*_pearson_h*.csv')
dtw_normal = []
df = pd.DataFrame()
for file in dtw_normal_files:
    print(file)
    df[file] = pd.read_csv(file, header=None)
i = 1
for col in df.columns:
    df.rename(columns={col:"run " + str(i)}, inplace=True)
    i +=1
df['epoch'] = range(1,51)
dfm_1 = df.melt('epoch', var_name='cols', value_name='DTW')
dfm_1['experiment'] = "missing_0.1"

os.chdir("/Users/alainamahalanabis/Documents/cycleGAN/missing_V_0.2")
dtw_normal_files = glob.glob('*_pearson_h*.csv')
dtw_normal = []
df = pd.DataFrame()
for file in dtw_normal_files:
    print(file)
    df[file] = pd.read_csv(file, header=None)
i = 1
for col in df.columns:
    df.rename(columns={col:"run " + str(i)}, inplace=True)
    i +=1
df['epoch'] = range(1,51)
dfm_2 = df.melt('epoch', var_name='cols', value_name='DTW')
dfm_2['experiment'] = "missing_0.2"

os.chdir("/Users/alainamahalanabis/Documents/cycleGAN/missing_V_0.4")
dtw_normal_files = glob.glob('*_pearson_h*.csv')
dtw_normal = []
df = pd.DataFrame()
for file in dtw_normal_files:
    print(file)
    df[file] = pd.read_csv(file, header=None)
i = 1
for col in df.columns:
    df.rename(columns={col:"run " + str(i)}, inplace=True)
    i +=1
df['epoch'] = range(1,51)
dfm_4 = df.melt('epoch', var_name='cols', value_name='DTW')
dfm_4['experiment'] = "missing_0.4"

os.chdir("/Users/alainamahalanabis/Documents/cycleGAN/missing_V_0.5")
dtw_normal_files = glob.glob('*_pearson_h*.csv')
dtw_normal = []
df = pd.DataFrame()
for file in dtw_normal_files:
    print(file)
    df[file] = pd.read_csv(file, header=None)
i = 1
for col in df.columns:
    df.rename(columns={col:"run " + str(i)}, inplace=True)
    i +=1
df['epoch'] = range(1,51)
dfm_5 = df.melt('epoch', var_name='cols', value_name='DTW')
dfm_5['experiment'] = "missing_0.5"


dfm = pd.concat([dfm_1, dfm_2, dfm_4, dfm_5], axis=0)
```

```python
#sns.set_style("whitegrid")
g=sns.catplot(x="epoch", y="DTW", hue='experiment', legend = False, data=dfm, kind='point', 
              height=8.27, aspect=15.7/8.27, ci=None)
g.ax.set_xlabel(xlabel = "epoch", fontsize=30)
g.ax.set_ylabel(ylabel = "Pearson Correlation Coefficient", fontsize=30)
g.ax.legend(fontsize=30)
g.ax.set_title("Pearson correlation vs. epoch Imputation", fontsize=30)
g.ax.tick_params(labelsize=15)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=30)
```

```python
os.chdir("/Users/alainamahalanabis/Documents/cycleGAN/figures/")
g.figure.set_size_inches(20, 20)
g.figure.savefig('imputation_pearson.png')
```
